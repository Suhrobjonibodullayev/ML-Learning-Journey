```python
numbers = {1,2,3,4,5,3,2,1,3,3,5} # sonlar 1 martadan saqlanadi.

mevalar = set() # bo'sh to'plam

mevalar.add('qulupnay')
mevalar.update(['apelsin', 'kiwi' , 'gilos' , 'anor' , 'olma' , 'malina'])
mevalar.discard('gilos')
mevalar.remove('olma')

A = {1,2,3,4}
B = {3,4,5,6}

print(A|B)  # 2 to'plamni birlashtirish
print(A.union(B))

print(A&B) # 2 to'plam kesishmasini (bir xil elementlarini) topish
print(A.intersection(B))

print(A-B) # A da bor B da yo'q
print(B.difference(A))  # B da bor A da yo'q

print(A^B)
print(A.symmetric_difference(B))
```

### **Nesting** -- Bir ma'lumot turi ichida boshqa ma'lumot turini saqlash.

```python
car1 = {
        'model':'nexia 3',
        'rang':'qora',
        'yil':2015,
        'narh':9000,
        'km':89000,
        'korobka':'mexanika'
        }

car2 = {
        'model':'gentra',
        'rang':'qizil',
        'yil':2019,
        'narh':15000,
        'km':20000,
        'korobka':'mexanika'
        }

cars = [car1, car2] # lug'atlar ro'yxati , ro'yxat ichida lug'at

print(cars[0]['model'])


malibus=[] # Malibu mashinalari uchun bo'sh ro'yxat yaratdik
for n in range(10):
    new_car = { # har bir yangi mashina uchun lug'at yaratamiz
        'model':'malibu',
        'rang':None, # rangi noaniq
        'yil':2020,
        'narh':None, # narhi belgilanmagan
        'km':0,
        'korobka':'avto'
        }
    malibus.append(new_car) # yangi lug'atni ro'yxatga qo'shamiz

for malibu in malibus[:3]:
    malibu['rang']='qizil'
for malibu in malibus[3:6]:
    malibu['rang']='qora'
for malibu in malibus[6:]: # ohirgi 4 ta mashinani
    malibu['rang']='qora'  # rangi qora
    malibu['korobka']='mexanika' # korobkasi mexanika

for malibu in malibus:
    if malibu['korobka']=='avto':
        malibu['narh']=40000
    else:
        malibu['narh']=35000
```

## While loop

While (toki) shart bajarilarkan takrorlanish davom etadi.

For ---> Qancha marta aylanish oldindan ma'lum bo'lsa ishlatiladi.

While --> Qachon tugashi oldindan noma'lum bo'lsa ishlatiladi.

loop ni to'xtatishni bir qancha yo'llari bor:
1) foydlanuvchiga berish.
2) flag (ishora) orqali.
3) break operatori bilan to'xtatish mumkin.

Break operatori - ma'lum shart bajarilganda to'xtatish
Continue operatori - ma'lum shart bajarilganda tashlab o'tish.

```python
son = 1

while son <= 5:
  print(son, end=' ')
  son +=1

savol = "istalgan sonni kiriting , dasturni to'xtatish uchun 'exit' deb yozing "
qiymat = ''

while qiymat != 'exit':
  qiymat = input(savol)
  if qiymat != 'exit':
    print(float(qiymat) ** 2)
```

```python
savol = "istalgan sonni kiriting , dasturni to'xtatish uchun 'exit' deb yozing "
qiymat = ''
flag = True # flag orqali to'xtatish

while flag: # toki flag true ekan
  qiymat = input(savol)
  if qiymat == 'exit':
    flag = False
  else:
    print(float(qiymat) ** 2)
```

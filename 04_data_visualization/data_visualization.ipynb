plt.grid() Barcha line xususiyatlaridan foydalanish mumkin

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x) # sin(x)
cos = np.cos(x) # cos(x)

plt.figure(figsize=(10,4)) # rasmning hajmi
# sin(x) grafigi
plt.plot(x, sin,
         linewidth=1.5, # chiziq qalinligi (0.5 dan 3 gacha)
         linestyle='-.', # chiziq shakli (-) oddiy chiziq , (--) yuliq chiziq , (-.) chiziq va nuqta
         color='#8E44AD', # b'-(blue) 'c'-(cyan) 'g'(green)'k'(black) 'm'(magenta) 'r'(red) 'y'(yellow)
         alpha=0.7, # chiziq shaffofligi (0 shaffof, 0.5 yarim-shaffof, 1 shaffof emas)
         marker='h') # '.' - nuqta 'o' - aylana 'v' - pastga qaragan uchburchak '^' - tepaga qaragan uchburchak '*' - yulduz

# cos(x) grafigi
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color=(44/255, 144/255, 245/255),
         alpha=0.7,
         marker='v',
         mec = 'blue',
         mew = 2,
         mfc = 'yellow',
         ms = 10
         )

# Grafik nomi:
plt.title('sin(x) funksiya grafigi')

# x - o'q nomi:
plt.xlabel("pi")

# y - o'q nomi:
plt.ylabel("sin(x)")

# Har bir chiziq haqida ma'lumot:
plt.legend(['sin(x)','cos(x)'], loc=0, fontsize='x-large')

# plt.grid() defoult # line xususiyatlaridan foydalanish mumkin
plt.grid(axis = 'x', alpha = 0.3, color = 'blue', lw = 2, ls = ':')

plt.show()

# Rasmni saqlab olamiz:
plt.savefig('sin-cos-grafik.png')

# Saqlangan rasmni ochish
# from IPython.display import Image
# Image('sin-cos-grafik.png')
```

**melt** funksiyasi jadvlani eniga uzundan bo'yiga uzun holatga olib kelishda qo'llaniladi. Bunda df dagi bir nechta ustunlarni yagona variable ustuniga keltirish mumkin. Qiymatlar esa har bir variable uchun value ustunida ko'rsatiladi.

Guruhlangan ustunli grafik chizish uchun **catplot** metodidan foydalanamiz. catplot yordamida turli grafiklar chizish mumkin, shuning **kind='bar'** parametri yordamida aynan bar grafik chizmoqchi ekanmizni ko'rsatamiz.
So'ngra, x va y o'qi qiymatlarini ko'rsatamiz va eng muhimi guruhlash qaysi ustun bo'yicha bo'lishini ko'rsatish uchun **hue='variable'** parametridan foydalanmiz

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/flight_delays.csv")

data = pd.melt(df, id_vars=['Month'], value_vars=['AA','MQ']) # .melt() yordamida istalgancha ajratib olish mumkin:
sns.catplot(data=data, kind="bar", x="Month", y="value", hue="variable")
plt.show()
```

## **Heatmap**

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0, parse_dates=[0], usecols=[0,1])
df['month'] = [i.month for i in df.index]
df['year'] = [i.year for i in df.index]

df_m = df.groupby(['month','year']).max().unstack(level=0)

plt.figure(figsize=(16,6))

sns.heatmap(data=df_m,
            cmap="Greens",
            linewidths=1,
            annot=False, # annot	Kataklarda raqamni yozish (True yoki False)
            square=True)

plt.xlabel("Airline")

xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(12) + .5, labels=xticks_labels) # x o'qi ustuunlarni nomlash
plt.show()
```

## **Distribution - taqsimot**

Taqsimot grafik - o'zgaruvchi (ustun)da qanday qiymatlar mavjudligi va ularning ehtimolligini ko'rsatuvchi grafik.

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df=df[df.year>=2010] # 2010-yildan keyin chiqqan mashinalarni saqlab qolamiz
df=df[df.transmission!='Other'] # Faqat avtomat, mexanik va yarim-avtomat mashinalarni olib qolamiz
df=df[df.fuelType!='Other'] # Faqat benzin, dizel va gibrid mashinalarni olib qolamiz
```

**displot**-turli taqsimotlarni jamlaydigan funksiya

displot yoki histplot ikkisining farqi histplot kdeplot, ecdfplot funksiyalari yordamida chizmaga ko'proq o'zgartirishlar kiritish mumkin.

```python
#plt.figure(figsize=(6,3))

# hue yordamida 2 - parametr qo'shiladi \ multiple="stack" ranglarni ajratib ko'rsatadi ustama ust
sns.displot(data=df,
            x="mileage",
            hue="transmission",
            multiple="stack",
            height=3, aspect=2, # figsize orniga height va aspect beriladi
            )

plt.show()
```

```python
# col parametri yordamida har bir grafikni ajratib chizish mumkin:
sns.displot(data=df,
            x="price",
            col="transmission",
            height=3, aspect=1,
            kde = True,
            color='crimson',       # Ustunlar rangi (chiroyli ko'k)
            edgecolor='#278a4e',     # Ustunlar orasidagi oq chegara
            linewidth=1.2
            )
plt.show()
```

```python
df = sns.load_dataset("penguins") # datasetni yuklash

sns.histplot(data = df, x="bill_length_mm",
         hue="species",
         bins = 30,
         kde = True,
         edgecolor='crimson',     # Ustunlar orasidagi oq chegara
         linewidth=1.2)         # Ustunlar qalinligi

plt.title("Tumshug' uzunligi taqsimoti", color="b")
plt.grid(color = 'grey', linewidth = 0.2)
plt.show()
```

```python
fig = px.scatter(
    iris,
    x="sepal_length",
    y="petal_length",
    color="species", # Hue bilan bir xil Guruxlash va alohida ranglar bilan
    size= 'sepal_width' , # 'sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species', 'species_id'
    size_max=16, # nuqta hajmi max qancha bo'lishi mumkinligini chegarasi.
    title="Iris Dataset", # sarlavha
    template='plotly_dark', # orqa fonni dark qilish
    #text = "species", 'sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species', 'species_id'
    color_discrete_map={
        "setosa":"red",
        "versicolor":"green",
        "virginica":"blue"
    },
    category_orders={
        "species": ["setosa", "versicolor", "virginica"]
    },
    labels={
        "sepal_width": "Sepal Width (cm)",
        "petal_width": "Petal Width (cm)"
    },
    symbol= 'species', # Shakl bilan guruxga ajraladi.
    trendline= "ols", # line chizish 'lowess', 'rolling', 'ewm', 'expanding', 'ols'
    trendline_color_override="yellow",
    trendline_scope="trace",
    #facet_col="species", # Grafikni guruhlash uchun
    width=900, # eni
    height=600 # bo'yi
)

fig.show()
```

## Bar

| Parametr                 | Vazifasi                                                                 |
| ------------------------ | ------------------------------------------------------------------------ |
| `barmode`                | Ustunlarni joylashtirish usuli (`group`, `stack`, `overlay`, `relative`) |
| `base`                   | Har bir ustunning boshlanish nuqtasi                                     |
| `error_x`                | X o'qi bo'yicha xatolik (error bar)                                      |
| `error_x_minus`          | X o'qi uchun pastki xatolik                                              |
| `error_y`                | Y o'qi bo'yicha xatolik                                                  |
| `error_y_minus`          | Y o'qi uchun pastki xatolik                                              |
| `text_auto`              | Ustun ustiga qiymatlarni avtomatik yozish (`True`, `'.2f'`, va hokazo)   |
| `pattern_shape`          | Ustunlarga tekstura (naqsh) berish                                       |
| `pattern_shape_sequence` | Teksturalar ketma-ketligi                                                |
| `pattern_shape_map`      | Kategoriyaga qarab tekstura berish                                       |


```python
fig = px.bar(
    tips,
    x="sex",
    y="tip",
    template='plotly_dark',
    color_discrete_sequence = ["crimson"],
    barmode="overlay", # Ustunlarni joylashtirish usuli (group, stack, overlay, relative)
    # base=0, # Har bir ustunning boshlanish nuqtasi
    text_auto=True, # Ustun ustiga qiymatlarni avtomatik yozish (True, '.2f', va hokazo)
    #error_x=0.2, # X o'qi bo'yicha xatolik (error bar)
    #error_y=0.2, # Y o'qi bo'yicha xatolik
    pattern_shape="sex", # Ustunlarga tekstura (naqsh) berish - ustun nomi beriladi
    pattern_shape_sequence=['/', 'x'], # Teksturalar ketma-ketligi
    pattern_shape_map={
        "Male": "/",
        "Female": "x"
    } # Kategoriyaga qarab tekstura berish
)
fig.show()
```

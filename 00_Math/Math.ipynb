````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
sns.set_theme()

# ── 1. Normal taqsimot yaratish ─────────────
mu    = 170   # o'rtacha bo'y (sm)
sigma = 10    # standart og'ish

x = np.linspace(130, 210, 300)    # x o'qi
y = stats.norm.pdf(x, mu, sigma)  # PDFy


# ── 2. Grafik ───────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# PDF
axes[0].plot(x, y, color='#2ECB9B', linewidth=2.5)
axes[0].fill_between(x, y, alpha=0.3, color='#2ECB9B')
axes[0].axvline(mu, color='red', linestyle='--')
axes[0].set_title('Normal PDF — Bo\'y taqsimoti')

# # Simulatsiya
samples = np.random.normal(mu, sigma, 1_000)
axes[1].hist(samples, bins=20, color='#2ECB9B', edgecolor='white')
axes[1].axvline(mu, color='red', linestyle='--')
axes[1].axvline(mu+sigma, color='black')
axes[1].axvline(mu-sigma, color='black')
axes[1].set_title('1000 ta tasodifiy namuna')
plt.tight_layout() ; plt.show()
````

1. Binomial Distribution nima?

Binomial Distribution — ma'lum sondagi mustaqil tajribalarda ma'lum natija necha marta sodir bo'lish ehtimolini hisoblaydigan taqsimot.

Misollar:

10 ta lampadan nechta ishlaydi. 1 yoki 0

20 ta urug'dan nechta unadi. Success yoki Failure

15 ta odamdan nechta Fanta Orange tanlaydi. Orange yoki Grape

Har bir tajribada faqat 2 ta natija bo'lishi kerak.

parametrlar P(muvaffaqiyat ehtimoli) va n(nechta marta)

tanga misolida n=10 10 marta tashlandi p=0.5 chikka tushishi 50 %

$$
P(X=k)=\binom{n}{k}p^k(1-p)^{n-k}
$$

n -> Jami tajribalar soni.

k -> Success soni.

p -> Bitta tajribada success ehtimoli.

n−k -> Failure soni.

$$
\binom{n}{k}=\frac{n!}{k!(n-k)!}
$$

"n ta ichidan k tasini tanlash usullari."

Binomial Distribution ishlashi uchun shartlar

Faqat 2 ta natija bo'lishi kerak. Success / Failure . Ha / Yo'q

Tajribalar soni oldindan ma'lum. Masalan n = 20

Har bir tajriba mustaqil.

Har bir tajribada success ehtimoli o'zgarmaydi. Masalan har safar p = 0.5

````python
 # ── 1. Parametrlar ──────────────────────────
n = 10    # tanga necha marta tashlanadi
p = 0.5   # 'Chikka' chiqish ehtimoli

# ── 2. PMF — har k uchun ehtimollik ────────
k = np.arange(0, n+1)         # 0 dan 10 gacha
# print(k)

# print(n)
# print(k)
pmf = stats.binom.pmf(k, n, p)
# print(pmf)


# print(f'P(5 Chikka) = {stats.binom.pmf(5,n,p):.3f}')  # 0.246

# # # ── 3. Grafik ───────────────────────────────
fig, axes = plt.subplots(1,2, figsize=(14, 5))

# PMF Bar chart
axes[0].bar(k, pmf, color='#3B82F6', edgecolor='red')
axes[0].set_title('Binomial PMF (n=10, p=0.5)')

# # n=10, p=0.3 bilan solishtirish
pmf2 = stats.binom.pmf(k, n, 0.3)
axes[1].bar(k-0.2, pmf,  width=0.4, label='p=0.5', color='#3B82F6')
axes[1].bar(k+0.2, pmf2, width=0.4, label='p=0.3', color='red')
axes[1].legend() ; axes[1].set_title('p=0.5 vs p=0.3')
plt.tight_layout() ; plt.show()
````

Standard Deviation --> Variance birlikni o'zgartirib yuboradi.

Masalan bo'y -> cm / Variance -> cm²

O'rtacha qiymatdan odatda qancha uzoqlikda joylashganini.

σ = √Variance

Bessel's Correction

s² = Σ(x−x̄)² / (n−1)

Sample mean o'z sample markaziga juda yaqin bo'ladi. Natijada farqlar kichikroq chiqadi. Variance ham kichrayadi. Agar n ga bo'lsak variance doim past baholanadi. Buni tuzatish uchun n−1 ishlatiladi.

| Formula            | Population       | Sample                |
| ------------------ | ---------------- | --------------------- |
| Mean               | μ = Σx / N       | x̄ = Σx / n           |
| Variance           | σ² = Σ(x−μ)² / N | s² = Σ(x−x̄)² / (n−1) |
| Standard deviation | σ = √σ²          | s = √s²               |


````python
import numpy as np

x = np.array([3, 13, 19, 24, 29])

# Sample mean
np.mean(x)

# Sample variance (n-1)
np.var(x, ddof=1)

# Sample std
np.std(x, ddof=1)
````

ddof=0 (standart) → population variance/std.

ddof=1 → sample variance/std, ya'ni n-1 ga bo'linadi.

````python
import numpy as np
import pandas as pd

data = [2, 4, 4, 4, 5, 5, 7, 9]
# Populyatsiya dispersiyasi (ddof=0, standart)
print(np.var(data))             # 4.0
# Namunaviy dispersiya (ddof=1, Bessel korreksiyasi)
print(np.var(data, ddof=1))     # 4.571428...
# Pandas bilan (ddof=1 standart)
s = pd.Series(data)
print(s.var())                  # 4.571428...
print(s.var(ddof=0))            # 4.0  (populyatsiya)
````

Korrelyatsiya ≠ Sabab-natija! 'Correlation does not imply causation'

Musbat va manfiy corr(bog'liqlik) -> r > 0: X↑ → Y↑ (musbat). r < 0: X↑ → Y↓ (manfiy). r ≈ 0: bog'liqlik yo'q

Standard Error (SE) — bu sample meanning qanchalik o'zgarishini o'lchaydigan kattalik.

Agar populationdan qayta-qayta sample olsangiz, har safar sample mean bir oz farq qiladi. Standard Error ana shu farqning o'rtacha miqdorini bildiradi.

o'rib turganingizdek, sample mean har safar ozgina o'zgaradi.

Mana shu o'zgarishning standart og'ishi Standard Error deyiladi.

Population standard deviation ma'lum bo'lsa

$$
SE = \frac{\sigma}{\sqrt{n}}
$$


σ — population standard deviation

n — sample size

Population standard deviation noma'lum bo'lsa

$$
SE = \frac{s}{\sqrt{n}}
$$

s — sample standard deviation.

| Standard Deviation (SD)                      | Standard Error (SE)                                |
| -------------------------------------------- | -------------------------------------------------- |
| Ma'lumotlarning tarqalishini o'lchaydi       | Sample meanlarning tarqalishini o'lchaydi          |
| Population yoki sample qiymatlariga tegishli | Faqat sample statistic (masalan, mean) ga tegishli |
| Formula: σ yoki s                            | Formula: σ/√n yoki s/√n                            |


Central Limit Theorem shuni aytadi:
Ko'p sample olsangiz, sample meanlar normal taqsimlanadi.

$$
\bar{X} \sim N\left(\mu,\frac{\sigma}{\sqrt{n}}\right)
$$

H₁: o'zgarish bor.

Alternative Hypothesis (H₁) — bu siz isbotlamoqchi bo'lgan da'vo.

Null Hypothesis (H₀) esa "hech qanday o'zgarish yo'q" degan boshlang'ich taxmin.

Eski batareyaning o'rtacha ishlash vaqti:
100 soat.

Kompaniya yangi batareya uzoqroq ishlaydi deb da'vo qilmoqda.

H₀: μ = 100
H₁: μ > 100

H₀: yangi batareya farq qilmaydi.

H₁: yangi batareya yaxshiroq.

Ikki tomonlama (Two-tailed)

Siz faqat farq borligini tekshirasiz.
H₀: μ = 100
H₁: μ ≠ 100

Bu "ko'paygan ham bo'lishi mumkin, kamaygan ham" degani.

O'ng tomonlama (Right-tailed)

Faqat oshganini tekshirasiz.

H₀: μ = 100
H₁: μ > 100

Misol:
Yangi dori bemorlarni tezroq sog'aytiradimi?

Chap tomonlama (Left-tailed)

Faqat kamayganini tekshirasiz.

H₀: μ = 100
H₁: μ < 100

Misol:
Yangi dvigatel yoqilg'i sarfini kamaytirdimi?

| Siz tekshirayotgan savol | Alternative Hypothesis |
| ------------------------ | ---------------------- |
| Farq bormi?              | H₁: μ ≠ μ₀             |
| Kattaroqmi?              | H₁: μ > μ₀             |
| Kichikroqmi?             | H₁: μ < μ₀             |


P-value — Null Hypothesis to'g'ri bo'lgan sharoitda kuzatilgan yoki undan ham kuchliroq natija tasodifan yuz berish ehtimoli.

Agar aslida ikkala dori bir xil bo'lsa,

Juda kichik. 0.001.

Eng mashhur chegara Statistikada odatda α = 0.05 ishlatiladi.

Agar p < 0.05 bo'lsa Null Hypothesis rad qilinadi.

Agar p ≥ 0.05 bo'lsa Null Hypothesis rad qilinmaydi.

Faraz qilaylik aslida ikkala dori bir xil. Siz tajribani 100 marta takrorladingiz. 0.05 degani 100 ta tajribaning taxminan 5 tasida aslida farq bo'lmasa ham siz "farq bor" deb noto'g'ri xulosa qilishingiz mumkin. Bu **Type I Error** yoki **False Positive** deb ataladi.

P-value farqning kattaligini o'lchamaydi. U faqat farqning statistik ishonchliligini ko'rsatadi.

| Haqiqiy holat | Qaror: H₀ rad etilmadi                                                                 | Qaror: H₀ rad etildi                                                                |
| ------------- | -------------------------------------------------------------------------------------- | ----------------------------------------------------------------------------------- |
| **H₀ rost**   | To'g'ri qaror. Ishonch darajasi = **1 − α**                                            | **False Positive (I tur xato)**. Yo'q narsani "bor" deb topdik. Ehtimolligi = **α** |
| **H₁ rost**   | **False Negative (II tur xato)**. Haqiqiy farqni "ko'rmay qoldik". Ehtimolligi = **β** | To'g'ri qaror. **Statistik kuch (Power) = 1 − β**                                   |


| Tushuncha                    | Formula   | Ma'nosi                                               |
| ---------------------------- | --------- | ----------------------------------------------------- |
| Ishonch darajasi             | **1 − α** | H₀ rost bo'lganda uni noto'g'ri rad etmaslik ehtimoli |
| I tur xato (False Positive)  | **α**     | Yo'q effektni bor deb topish                          |
| II tur xato (False Negative) | **β**     | Bor effektni aniqlay olmaslik                         |
| Statistik kuch (Power)       | **1 − β** | Haqiqiy effektni to'g'ri aniqlash ehtimoli            |


**Z test**

Namuna o'rtachasini ma'lum populyatsiya o'rtachasi bilan solishtirish uchun ishlatiladigan statistik test. Populyatsiya standart chetlanishi (σ) oldindan ma'lum bo'lganda qo'llaniladi.

Standart chetlanish oldindan berilgan yoki aniq hisoblangan bo'lishi kerak

Odatda n ≥ 30, Markaziy Limit Teoremasi kuchga kiradi

Ma'lumot normal yoki n katta bo'lgani uchun taqsimot normalga yaqin

### Z-test formulalari

### 1. Z-test formulasi

$$
Z=\frac{\bar{x}-\mu_0}{\sigma/\sqrt{n}}
$$

Bu yerda:

- $Z$ — test statistikasi
- $\bar{x}$ — sample mean
- $\mu_0$ — null hypothesis dagi o'rtacha qiymat
- $\sigma$ — population standard deviation
- $n$ — sample size

---

### 2. Standard Error (SE)

$$
SE=\frac{\sigma}{\sqrt{n}}
$$

Shunda Z-test formulasi:

$$
Z=\frac{\bar{x}-\mu_0}{SE}
$$

---

### Qaror qoidalari

### Right-tailed test

Alternative hypothesis:

$$
H_1:\mu>\mu_0
$$

Reject $H_0$, agar

$$
Z>Z_\alpha
$$

---

### Left-tailed test

Alternative hypothesis:

$$
H_1:\mu<\mu_0
$$

Reject $H_0$, agar

$$
Z<-Z_\alpha
$$

---

### Two-tailed test

Alternative hypothesis:

$$
H_1:\mu\neq\mu_0
$$

Reject $H_0$, agar

$$
|Z|>Z_{\alpha/2}
$$

---

### Eng ko'p ishlatiladigan Critical Z qiymatlar

| Confidence Level | α | Two-tailed | One-tailed |
|------------------|---|------------|------------|
| 90% | 0.10 | ±1.645 | 1.282 |
| 95% | 0.05 | ±1.960 | 1.645 |
| 99% | 0.01 | ±2.576 | 2.326 |

````python
import numpy as np
from scipy import stats

mu0, sigma = 70, 10
x_bar, n = 74, 36

z = (x_bar-mu0)/(sigma/np.sqrt(n))
p = 2*(1-stats.norm.cdf(abs(z)))
# z = 2.40, p = 0.0164
````

**T test**

Populyatsiya standart chetlanishi (σ) noma'lum bo'lganda, uning o'rniga namuna standart chetlanishi (s) ishlatiladigan test. Kichik namunalarda ayniqsa muhim.

Faqat namuna standart chetlanishi (s) hisoblanadi va σ o'rniga ishlatiladi

Odatda n < 30 bo'lganda t-taqsimot normal taqsimotdan aniqroq natija beradi

df = n − 1 — t-taqsimot shakli namuna hajmiga bog'liq

### T-test formulalari

### 1. One Sample t-test

$$
t=\frac{\bar{x}-\mu_0}{s/\sqrt{n}}
$$

Bu yerda:

- $t$ — t-statistikasi
- $\bar{x}$ — sample mean
- $\mu_0$ — null hypothesis dagi o'rtacha qiymat
- $s$ — sample standard deviation
- $n$ — sample size

Degrees of Freedom (df):

$$
df=n-1
$$

---

### 2. Independent Two Sample t-test (Equal Variances)

$$
t=\frac{\bar{x}_1-\bar{x}_2}
{s_p\sqrt{\frac{1}{n_1}+\frac{1}{n_2}}}
$$

Pooled Standard Deviation:

$$
s_p=
\sqrt{
\frac{(n_1-1)s_1^2+(n_2-1)s_2^2}
{n_1+n_2-2}
}
$$

Degrees of Freedom:

$$
df=n_1+n_2-2
$$

---

### 3. Welch's t-test (Unequal Variances)

$$
t=
\frac{\bar{x}_1-\bar{x}_2}
{\sqrt{\frac{s_1^2}{n_1}+\frac{s_2^2}{n_2}}}
$$

Welch Degrees of Freedom:

$$
df=
\frac{\left(\frac{s_1^2}{n_1}+\frac{s_2^2}{n_2}\right)^2}
{\frac{\left(\frac{s_1^2}{n_1}\right)^2}{n_1-1}
+
\frac{\left(\frac{s_2^2}{n_2}\right)^2}{n_2-1}}
$$

---

### 4. Paired t-test

$$
t=
\frac{\bar{d}}
{s_d/\sqrt{n}}
$$

Bu yerda:

- $\bar{d}$ — juft kuzatuvlar farqining o'rtachasi
- $s_d$ — farqlarning standart og'ishi
- $n$ — juft kuzatuvlar soni

Degrees of Freedom:

$$
df=n-1
$$

---

### Qaror qoidalari

### Right-tailed test

$$
H_1:\mu>\mu_0
$$

Reject $H_0$, agar

$$
t>t_\alpha
$$

---

### Left-tailed test

$$
H_1:\mu<\mu_0
$$

Reject $H_0$, agar

$$
t<-t_\alpha
$$

---

### Two-tailed test

$$
H_1:\mu\neq\mu_0
$$

Reject $H_0$, agar

$$
|t|>t_{\alpha/2}
$$

````python
from scipy import stats

scores = [68,72,75,70,74,71,69,73,76,72]

t_stat, p = stats.ttest_1samp(
    scores, popmean=70
)

print(f't = {t_stat:.2f}, p = {p:.4f}')
# t = 1.84, p = 0.0989
````

````python
from scipy import stats

# Eski va yangi sayt dizaynidagi kunlik savdo (ming so'm)
eski  = [45, 42, 48, 44, 46, 43, 47, 45, 44, 46]
yangi = [48, 50, 47, 49, 51, 48, 50, 49, 52, 48]

t_stat, p_value = stats.ttest_ind(yangi, eski)
alpha = 0.05

if p_value < alpha:
    print('H0 rad etildi: farq ahamiyatli')
else:
    print('H0 qoldirildi: yetarli dalil yoq')

# Natija: t = 4.72, p = 0.0002  ->  H0 rad etildi
````

Multiple testing uchun tuzatish usullarini qo'llamaslik.

Oldini olish uchun:

Bu yerda:

- $v_1,v_2,\ldots,v_n$ — vektorlar

Rank — matritsaning nechta yangi yo'nalish (dimension) hosil qila olishini bildiradi.

Rank tushunchasi span va linear independence bilan chambarchas bog'liq.

Intuitiv tushuncha Tasavvur qiling, sizda quyidagi vektorlar bor:

v1 = [1,0] v2 =[0,1]

Bu ikki vektor biri o'ngga, biri yuqoriga qaragan.

Ikkalasi ham mustaqil.

Demak. Rank=2 Chunki ular ikkita turli yo'nalishni hosil qilyapti.

Rank = 0
Faqat nol vektor mavjud. Hech qanday yo'nalish yo'q.

Rank = 1

Faqat bitta yo'nalish. Span bitta chiziq.

v1 =[1,2] v2= [2,4]

Bu ikkisi bir xil chiziqda.

Rank = 2

Ikki mustaqil yo'nalish.

Rank = 3

Uchta mustaqil yo'nalish.

Rank va Span bog'lanishi
Rank aslida span ning o'lchami.

| Matritsa                              | Rank |
| ------------------------------------- | ---- |
| $\begin{bmatrix}1&0\0&1\end{bmatrix}$ | 2    |
| $\begin{bmatrix}1&2\2&4\end{bmatrix}$ | 1    |
| $\begin{bmatrix}0&0\0&0\end{bmatrix}$ | 0    |


Agar barcha ustunlar mustaqil bo'lsa

Rank=ustunlar soni

Agar ba'zilari boshqalardan hosil bo'lsa,
rank kamayadi.

Eng ko'p ishlatiladigan usul — Gaussian Elimination (Row Echelon Form).

Qadamlar:

Matritsani satrlar bo'yicha soddalashtiring.

Nol bo'lmagan satrlarni sanang.

Shu son rank bo'ladi.

````python
v1 = np.array([2, 4])
v2 = np.array([1, 2])
v3 = np.array([0, 5])

# 1) b span{v1,v2} ichidami
A = np.column_stack([v1, v2])
c, *_ = np.linalg.lstsq(
    A, [3, 6], rcond=None)
# xato = 0 -> ichida (chiziqda)

print(np.linalg.matrix_rank([v1,v2]))  #1
print(np.linalg.matrix_rank([v1,v3]))  #2
````

v3 = 2*v1 + 3*v2. Ya'ni v3 hech qanday yangi yo'nalish qo'shmaydi — ortiqcha. Dependent

Hech birini ikkinchisidan yasab bo'lmaydi. Har biri yangi yo'nalish — ikkalasi ham kerak. har biri yangi yo'nalish Independent

matrix_rank = mustaqil vektorlar soni. Agar rank = vektorlar soni bo'lsa — erkin.

det = 0 emas — erkin.  det = 0 — bog'liq. Faqat n×n uchun.

d o'lchamli fazoda d dan ko'p vektor bo'lsa — doim bog'liq (masalan 2D da 3 vektor).

Amalda «aniq 0» kam bo'ladi. det juda kichik bo'lsa — «deyarli bog'liq» degani (shovqinli data).

````python
M = np.array([
    [1, 0, 2],
    [0, 1, 3],
])   # ustunlar: v1, v2, v3

r = np.linalg.matrix_rank(M)
print("rank =", r)   # rank = 2
# rank(2) < ustunlar(3)  ->  BOG'LIQ

# kvadrat matritsa uchun determinant:
print(np.linalg.det([[1, 0], [0, 1]]))  # 1.0
````

````python
import pandas as pd
import numpy as np
from numpy.linalg import matrix_rank

# 1. Sun'iy dataset yaratish

df = pd.DataFrame({
    "A": [1, 2, 3, 4, 5],
    "B": [2, 4, 6, 8, 10],     # B = 2 * A
    "C": [5, 3, 6, 2, 7],
    "D": [6, 5, 9, 6, 12],     # D = A + C
    "E": [9, 8, 5, 7, 4],
    "F": [18, 16, 10, 14, 8]   # F = 2 * E
})

print("Original Dataset:")
print(df)

# 2. Matrix Rank

rank = matrix_rank(df.values)

print("\nUstunlar soni:", df.shape[1])
print("Matrix Rank:", rank)
print("Mustaqil ustunlar soni:", rank)

# 3. Bog'liq ustunlarni aniqlash

print("\nBog'liq ustunlar:")

print("B = 2 × A")
print("D = A + C")
print("F = 2 × E")

# 4. Ortiqcha ustunlarni olib tashlash


clean_df = df.drop(columns=["B", "D", "F"])

print("\nToza Dataset:")
print(clean_df)

print("\nYangi Matrix Rank:", matrix_rank(clean_df.values))
````

### Determinant Formula

2×2 matritsa:

$$
A=
\begin{bmatrix}
a & b\\

\end{bmatrix}
$$

Determinant:

$$
\det(A)=ad-bc

### Inverse Matrix Formula

Agar

$$
\det(A)\neq0
$$

bo'lsa, inverse mavjud.

$$
A^{-1}
=
\frac{1}{ad-bc}
\begin{bmatrix}
d & -b\\
-c & a
\end{bmatrix}
$$

yoki

$$
A^{-1}
=
\frac{1}{\det(A)}
\begin{bmatrix}
d & -b\\
-c & a
\end{bmatrix}
$$

---

$$
\det(A)\neq0
$$

$$
\det(A)=0
$$

bo'lsa,

$$

mavjud emas (Singular Matrix).

1. Binomial Distribution nima?

Binomial Distribution — ma'lum sondagi mustaqil tajribalarda ma'lum natija necha marta sodir bo'lish ehtimolini hisoblaydigan taqsimot.

Misollar:

10 ta lampadan nechta ishlaydi. 1 yoki 0

20 ta urug'dan nechta unadi. Success yoki Failure

15 ta odamdan nechta Fanta Orange tanlaydi. Orange yoki Grape

Har bir tajribada faqat 2 ta natija bo'lishi kerak.

parametrlar P(muvaffaqiyat ehtimoli) va n(nechta marta)

tanga misolida n=10 10 marta tashlandi p=0.5 chikka tushishi 50 %

$$
P(X=k)=\binom{n}{k}p^k(1-p)^{n-k}
$$

n -> Jami tajribalar soni.

k -> Success soni.

p -> Bitta tajribada success ehtimoli.

n−k -> Failure soni.

$$
\binom{n}{k}=\frac{n!}{k!(n-k)!}
$$

"n ta ichidan k tasini tanlash usullari."

Binomial Distribution ishlashi uchun shartlar

Faqat 2 ta natija bo'lishi kerak. Success / Failure . Ha / Yo'q

Tajribalar soni oldindan ma'lum. Masalan n = 20

Har bir tajriba mustaqil.

Har bir tajribada success ehtimoli o'zgarmaydi. Masalan har safar p = 0.5

**Poisson Distributio**n nima?

Poisson Distribution ma'lum vaqt yoki ma'lum hudud ichida hodisa necha marta sodir bo'lish ehtimolini hisoblaydi.

Masalan:

1 soatda nechta telefon qo'ng'irog'i keladi.

1 kunda nechta mijoz keladi.

1 daqiqada serverga nechta so'rov keladi.

Poisson Distribution qachon ishlatiladi?

Quyidagi shartlar bajarilganda ishlatiladi.

Hodisalar sanaladi.

Hodisalar mustaqil.

O'rtacha tezlik (rate) o'zgarmaydi.

Vaqt yoki masofa oralig'i aniq.

$$
P(X=x)=\frac{\mu^x e^{-\mu}}{x!}
$$

Bu yerda:

X — tasodifiy o'zgaruvchi.

x — biz hisoblayotgan qiymat.

μ yoki λ — o'rtacha hodisalar soni.

e=2.71828

x! — factorial.

Bir biznesga o'rtacha 12 ta mijoz keladi.
Toping: Aynan 8 ta mijoz kelish ehtimoli.

$$
P(X=8)=
\frac{12^8e^{-12}}{8!}
$$

```python
import numpy as np

np.random.seed(0)
grades = np.random.randint(40, 101, size=(10, 4))  # 10 talaba, 4 fan
subjects = ['Math', 'Physics', 'History', 'English']
normalized = grades / 100
mean_scores = np.mean(grades, axis=0, keepdims=True)
median_scores = np.median(grades, axis=0, keepdims=True)
total_scores = np.sum(grades, axis=1, keepdims=True)
best_subject_idx = np.argmax(grades, axis=1)
labels = np.where(grades >= 90, 'A',
         np.where(grades >= 80, 'B',
         np.where(grades >= 70, 'C',
         np.where(grades >= 60, 'D', 'F'))))

print("Baholar:\n", grades)
print("Normallashtirilgan:\n", normalized)
print("O‘rtacha baholar:\n", mean_scores)
print("Umumiy ballar:\n", total_scores)
print("Eng yaxshi fan indeksi:\n", best_subject_idx)
print("Baholar harfi:\n", labels)
```

```python
import numpy as np

# Har bir qatorda: [ID, Matematika, Fizika, Ingliz tili]
grades = np.array([
    [101, 78, 65, 80],
    [102, 45, 50, 48],
    [103, 90, 88, 92],
    [104, 60, 70, 55],
    [105, 30, 40, 35]
])
# Faqat baholarni ajratamiz (ID ni olib tashlaymiz)
scores = grades[:, 1:]

# Har bir talaba bo'yicha o'rtacha baho
avg_scores = scores.mean(axis=1, keepdims=True)

# Yangi ustun sifatida qo'shamiz
grades_with_avg = np.concatenate((grades, avg_scores), axis=1)
print(grades_with_avg)

# O'rtacha baho 60 dan yuqori bo'lsa - Pass, aks holda - Fail
status = np.where(avg_scores >= 60, "Pass", "Fail")

# Talaba ID + Status ko'rinishida chiqaramiz
result = np.concatenate((grades[:, :1], status), axis=1)
print(result)

subject_avg = scores.mean(axis=0)
print("Fanlar bo'yicha o'rtacha baholar:", subject_avg)
```

```python
import numpy as np

# Ma'lumotlar
days = np.array(["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
distances = np.array([5.2, 6.0, 0.0, 5.5, 4.8, 7.2, 3.0])  # km
times = np.array([30, 35, 0, 32, 28, 40, 20])  # minutes

# 1. O'rtacha tezlik (km/min)
speeds = np.divide(distances, times, out=np.zeros_like(distances), where=times!=0)

# 2. 5 km dan ko'p yugurilgan kunlar
long_runs = np.where(distances > 5)[0]

# 3. Shape saqlab, har kunlik km/soat ni topamiz
speeds_kph = speeds * 60
speeds_kph = np.expand_dims(speeds_kph, axis=1)

# 4. Har bir kun holati: "Rest" (0 km), "Short" (≤5 km), "Long" (>5 km)
run_type = np.where(distances == 0, "Rest",
            np.where(distances > 5, "Long", "Short"))

# 5. Haftalik umumiy statistika
total_distance = distances.sum()
avg_speed = np.mean(speeds_kph[distances > 0])  # faqat yugurilgan kunlar
max_day = days[np.argmax(distances)]

# Natijalar
print("Tezliklar (km/soat):\n", speeds_kph)
print("5km+ kunlar:", days[long_runs])
print("Har kun holati:", run_type)
print("Umumiy masofa:", total_distance, "km")
print("O‘rtacha tezlik:", round(avg_speed, 2), "km/h")
print("Eng ko‘p yugurilgan kun:", max_day)
```

1. NumPy nima?
Python uchun tez, moslashuvchan massiv kutubxonasi.
Asosiy obyekti – ndarray (N-dimensional array).
C asosida yozilgan, juda tez ishlaydi.

2. ndarray va Python ro'yxati (list) o'rtasidagi farq?

| xususiyat        | `ndarray`                | `list`                  |
| ---------------- | ------------------------ | ----------------------- |
| Hajm             | Ruxsat etilmaydi         | Har xil o'lchamli       |
| Tezlik           | Tez (C bilan bog'langan) | Sekin (for loop orqali) |
| Vektorli amallar | Mavjud                   | Yo'q (loop kerak)       |
| Xotira           | Kam foydalanadi          | Ko'proq foydalanadi     |

3. Broadcasting nima?
Turli o'lchamdagi arraylar ustida amallar bajarish usuli.
Masalan:
a = np.array([[1], [2], [3]])
b = np.array([10, 20, 30])
print(a + b)

4. axis parametri nima qiladi?
Operatsiyani qaysi yo’nalishda bajarilishini belgilaydi.

a.sum(axis=0)  # ustun bo'yicha
a.sum(axis=1)  # qator bo'yicha

5. np.where qanday ishlaydi?
Shartga qarab qiymat tanlaydi:

a = np.array([1, 6, 3])
np.where(a > 5, 'High', 'Low')

6. keepdims=True nima qiladi?
Operatsiyadan so'ng shaklni saqlaydi (broadcast uchun qulay):

a = np.array([[1, 2], [3, 4]])
a.sum(axis=1, keepdims=True)

7. ufunc nima?
Universal funksiyalar, vektorli amallarni tez bajaradi.
Misol: np.sqrt([1, 4, 9])  # [1. 2. 3.]

8. reshape, ravel, flatten farqi nima?
reshape: yangi shaklga o‘zgartiradi (ko‘pincha view).
ravel: tekislash (view, imkon bo‘lsa).
flatten: tekislash (yangi nusxa yaratadi).

9. NaN bilan ishlash qanday?

np.isnan(a)
np.nanmean(a)
np.nan_to_num(a)

10. Indexing turlari:
Boolean indexing:

a = np.array([1, 2, 3, 4])
a[a > 2]  # [3 4]

Fancy indexing:

a[[0, 2]]  # [1 3]
